```python
import imgviz
from PIL import Image
import numpy as np
import os
```

```python
"""
"_background_": 0,      "face":1,
"hair":2,               "eye":3,
"eyeblow":4,            "nose":5,
"mouth":6,              "clothes":7,
"body":8,
"""
MASKS_PRIORITY = [7, 8, 1, 2, 3, 5, 6]
MASKS_INPUT_PATH = r"F:\WarehouseOfRui\unet-nested-multiple-classification-master\data\test\output\v3_92"
MASKS_OUTPUT_PATH = r"F:\WarehouseOfRui\FYP\Cropped_imgs\MASKS\OUTPUT_v3"
```

```python
def trans_multi_msk_to_single(input_p, msk_name):  #输入的是包括多个单个类mask的文件夹路径
    initial_msk = np.zeros([256,256], dtype=np.int32)
    masks = os.listdir(input_p)
    saved_p = os.path.join(MASKS_OUTPUT_PATH, msk_name + '_test.png')
    for idx in MASKS_PRIORITY:
        single_m = np.asarray(Image.open(os.path.join(input_p, masks[idx])), dtype=np.int32)
        for x in range(256):
            for y in range(256):
                if single_m[x][y] > 0:
                    initial_msk[x][y] = idx
    save_colored_mask(saved_p, initial_msk)

def save_colored_mask(save_path, mask):
    lbl_pil = Image.fromarray(mask.astype(np.uint8), mode="P")
    colormap = imgviz.label_colormap()
    lbl_pil.putpalette(colormap.flatten())
    lbl_pil.save(save_path)
```

```python
MASKS_FLODER_LI = os.listdir(MASKS_INPUT_PATH)
for folder in MASKS_FLODER_LI:
    folder_p = os.path.join(MASKS_INPUT_PATH, folder)
    trans_multi_msk_to_single(folder_p, folder)
    print('[Success]', folder_p)
```
